```python


stop = ['%', 'address', 'accessibility', 'accounts', 'admin', 'billing', 'blog', 'boutique', 'care', 'career', 'careers', 'clientservices', 'community', 'concierge', 'contact', 'contacto', 'contactus', 'cs', 'custom', 'customer', 'customer.service', 'customercare', 'customerservice', 'customersupport', 'custservice', 'dataprotection', 'design', 'dmca', 'ecommerce', 'enquiries', 'events', 'feedback', 'gdpr', 'hello', 'help', 'helpdesk', 'hi', 'hr', 'influencers', 'info', 'information', 'internetsales', 'it', 'jobs', 'kontakt', 'legal', 'mail', 'mail', 'mailorder', 'marketing', 'media', 'office', 'online', 'order', 'orders', 'pr', 'press', 'privacy', 'concerns', 'questions', 'recall', 'recruitment', 'repairs', 'returns', 'rsvp', 'sales', 'schools', 'service', 'shop', 'social', 'store', 'submissions', 'support', 'team', 'unsubscribe', 'us', 'warranty', 'web', 'webmaster', 'websales', 'webshop', 'wecare', 'weddings', 'wholesale']


twitter_junk = [
    '',
    
]

telephone_regex = '\d{3}(-)?\d{3}(-)?\d{4}'

# Get ready to store the information in a proper way
mailing_list = pd.DataFrame(columns = ['Name', 'Email', 'Domain', 'Ranking'])

gmail_re = re.compile('')

def gmail_correction(mail: str):
    return re.sub('@gmai(i|1)', '@gmail', mail)

def name_correction(name: str):
    return re.sub('\d', '', name)

mailing = []
for i in range(len(shopify)):
    field = shopify.loc[i, "Emails"]
    field = re.sub('\d{3}(-)?\d{3}(-)?\d{4}', '', field).split(",")
    field = [i for i in field if i]
    for email in field:
        mail_parts = email.split("@")
       
        if not mail_parts[0].find('.'):
            try:
                terms = sym_spell.word_segmentation(mail_parts[0]).corrected_string
            except:
                print(f'Fail on  {email}')
                break
        else:
             terms = mail_parts[0].replace('.', ' ')
        if not any(x in terms for x in stop):
            parse = nlp(terms)
            for x in parse.ents:
                if x.label == 380:
                    # Add to the table
                    mail_entry = {
                        'Name': name_correction(" ".join([str(i) for i in parse.ents])).title(),
                        'Email': gmail_correction(email), #Emails -> Email
                        'Domain': shopify.loc[i, "Domain"],
                        'Ranking': shopify.loc[i, "DomainRanking"]
                    }
                    mailing_list = mailing_list.append(mail_entry, ignore_index = True)
                    break
```

```python
bad_seq1 = re.compile("[0-9]{4,}")
bad_seq2 = re.compile("(%)")
twitter = []
for field in shopify['Twitter']:
    field = field.split(",")
    for account in field:
        x = account.split("/")[-1]
        if x != '' and not bad_seq1.match(x) and not bad_seq2.match(x):
            twitter.append(x)
```

```python
# https://github.com/twintproject/twint

c = twint.Config()
for name in twitter:
    c.Pandas = True
    c.Username = name
    try:
        twint.run.Lookup(c)
    except:
        f'Break on {name}'
```
